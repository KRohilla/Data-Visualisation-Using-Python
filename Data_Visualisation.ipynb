**Task 3 Data Visualisation With Python**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
data = pd.read_csv("/content/tips.csv")
data.sample(5)
```

```python
# shape of the dataset
data.shape
```

```python
# checking the missing values and datatypes of the dataset
data.info()
```

```python
data.isnull().sum()
```

```python
# unique values in each column
data.nunique()
```

```python
# statistical info about the dataset --> for numerical columns
data.describe()
```

```python
# description --> for categorical columns
data.describe(include='object')
```

```python
data['smoker'].value_counts()
```

**Data Visualisation**

```python
data['day'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
data['sex'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
# mostly size = 3 giving the highest tip
plt.bar(data['size'], data['tip'])
plt.xlabel('Size')
plt.ylabel('Tip')
plt.title('Size vs Tip')
plt.show()
```

```python
plt.hist(data['total_bill'], bins=30)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# checking the correlation between tips and total_bill and also classify that which sex gives highest tip
sns.scatterplot(x = data['total_bill'], y = data['tip'], hue=data['sex'])
```

```python
sns.scatterplot(x = data['total_bill'], y = data['tip'],hue = data['smoker'])
```

```python
# frequency distribution of total bill and tips
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(data['total_bill'])
plt.subplot(1,2,2)
sns.distplot(data['tip'])
```

```python
sns.regplot(x = data['total_bill'], y = data['tip'])
```

```python
# adding the additional categorical data
sns.lmplot(x = 'total_bill', y = 'tip', data = data, hue = 'smoker')
```

```python
# checking the distribution of smoker and sex with the tips
sns.barplot(x = data['smoker'], y = data['tip'], hue=data['sex'])
```

```python
# checking which day and time, the highest tip
sns.barplot(x = data['day'], y = data['tip'], hue=data['time'])
```

```python
# the five number summary of numerical columns --> for outliers
sns.boxplot(data)
```

```python
sns.pairplot(data, kind='scatter', hue='day', palette='Set1')
```

```python
sns.pairplot(data, kind='scatter', hue='size', palette='Set2')
```
